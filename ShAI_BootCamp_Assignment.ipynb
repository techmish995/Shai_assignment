```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load your dataset
data = pd.read_csv('Salaries.csv')
data
```

```python
data.columns
```

```python
num_rows, num_cols = data.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)

# Check data types of each column
data_types = data.dtypes
print("\nData types of each column:")
print(data_types)

# Check for missing values in each column
missing_values = data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python

mean_salary = data['TotalPay'].mean()
median_salary = data['TotalPay'].median()
mode_salary = data['TotalPay'].mode().values[0]
min_salary = data['TotalPay'].min()
max_salary = data['TotalPay'].max()

# Determine the range of salaries
salary_range = max_salary - min_salary

# Calculate the standard deviation of salaries
std_salary = data['TotalPay'].std()

# Print the results
print("Mean salary:", mean_salary)
print("Median salary:", median_salary)
print("Mode salary:", mode_salary)
print("Minimum salary:", min_salary)
print("Maximum salary:", max_salary)
print("Salary range:", salary_range)
print("Standard deviation of salary:", std_salary)
```

```python
# the two columns are completely empty thats why i chose to drop columns
# Drop specific columns
columns_to_drop = ['Notes', 'Status']
data_cleaned = data.drop(columns_to_drop, axis=1)

# Print the cleaned dataset
print(data_cleaned)
```

```python
# Create a histogram of salaries
plt.figure(figsize=(10, 6))
plt.hist(data['TotalPay'], bins=30, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()

# Create a pie chart of department proportions
department_counts = data['JobTitle'].value_counts()
top_departments = department_counts.head(10)  # Select the top 10 departments

plt.figure(figsize=(8, 8))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Top 10 Departments')
plt.show()
```

```python
# Group the data by one or more columns and calculate summary statistics
grouped_data = data.groupby(['JobTitle', 'BasePay'])['TotalPay'].agg(['mean', 'median', 'min', 'max', 'std'])

# Print the summary statistics for each group
print(grouped_data)

# Compare the average salaries across different groups
average_salaries = data.groupby(['JobTitle', 'BasePay'])['TotalPay'].mean()
print(average_salaries)
```

```python
# Specify the columns for correlation analysis
column1 = 'TotalPay'  # Salary column
column2 = 'Benefits'  # Benefits column

# Calculate the correlation coefficient
correlation_coefficient = data[column1].corr(data[column2])

# Plot a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data[column1], data[column2], alpha=0.5)
plt.xlabel(column1)
plt.ylabel(column2)
plt.title('Scatter Plot: {} vs. {}'.format(column1, column2))
plt.show()

# Print the correlation coefficient
print('Correlation coefficient between {} and {}: {}'.format(column1, column2, correlation_coefficient))
```
